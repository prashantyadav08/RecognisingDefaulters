```python
def unistat(frame):
    outputframe = pd.DataFrame(columns = ["count", "missing", "unique", "dtype", "mean", "mode", "min", "25%", "median", "75%", "max", "std", "skew", "kurt"])
    for i in frame.columns:
        if (frame[i].dtype == "int64") or (frame[i].dtype == "float64"):
            outputframe.loc[i] = [frame[i].count(), frame[i].isnull().sum(), frame[i].nunique(), frame[i].dtype, frame[i].mean(), frame[i].mode().values[0], frame[i].min(), frame[i].quantile(0.25), frame[i].median(), frame[i].quantile(0.75), frame[i].max(), frame[i].std(), frame[i].skew(), frame[i].kurt()]
        else:
            outputframe.loc[i] = [frame[i].count(), frame[i].isnull().sum(), frame[i].nunique(), frame[i].dtype, "-", "-", "-", "-", "-", "-", "-", "-", "-", "-"]

    return outputframe  

unistat(data)
```

## Under Sampling

By using the under sampling  we are trying to reduce the points of maximum labels.

```python
from collections import Counter
Counter(y_train)
```

```python
from collections import Counter
from imblearn.under_sampling import NearMiss
ns = NearMiss(.8)
x_train_ns,y_train_ns = ns.fit_sample(x_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_ns)))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(x_train_ns,y_train_ns)
```

```python
y_pred=classifier.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

Note:
    
    From the above result we can clearly see that our under sampled model gives very poor result. So don't use under
